```python
from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X_normalized)
    kmeanModel.fit(X_normalized)
    distortions.append(sum(np.min(cdist(X_normalized, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_normalized.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

# Cluster 

```python
import pandas

import pylab as pl

from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

```python

variables = pandas.read_csv("/Users/courtneycampbell/Desktop/comodo_data/netflow.csv", nrows = 1000) 

Y = variables[['srcPackets']].head(300)

X = variables[['srcBytes']].head(300)
```

```python

pca = PCA(n_components=1).fit(Y)

pca_d = pca.transform(Y)

pca_c = pca.transform(X)
```

```python
kmeans=KMeans(n_clusters=2)

kmeansoutput=kmeans.fit(Y)

kmeansoutput

pl.figure('2 Cluster K-Means')

pl.scatter(pca_c[:, 0], pca_d[:, 0], c=kmeansoutput.labels_)

pl.xlabel('Source Packets')

pl.ylabel('Source Bytes')

pl.title('2 Cluster K-Means')

pl.show()
```

# Isolation Forest Anomaly 


```python
#Loading Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
%matplotlib inline
```

```python
#Selecting Numeric Fields To Be Used In Modelling
X = credit[['srcBytes','srcPackets','dstBytes','dstPackets']]
```

```python
#Loading Min Max Scaling and Creating New Data Frame
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_normalized = pd.DataFrame(scaler.fit_transform(X))
```

```python
#Assining Column Names
X_normalized.columns = ['srcBytes','srcPackets','dstBytes','dstPackets']
```

```python
#Implementing Algorithm
from sklearn.ensemble import IsolationForest
iforest = IsolationForest()
iforest.fit(X_normalized)
```

```python
#Predicting Anomaly Class and Anomaly Score
Anomaly_Class = iforest.predict(X_normalized)
Anomaly_Score = iforest.decision_function(X_normalized)
```

```python
#Creating New Columns Based on Anomaly Detection
X['Anomaly_Class'] = Anomaly_Class
X['Anomaly_Score'] = Anomaly_Score
```

```python
#Viewing Class Counts
#-1 is anomaly, 1 is normal
X.groupby('Anomaly_Class')[['Anomaly_Score']].count()
```

```python
#Histogram of Anomaly Scores By Anomaly Class
X['Anomaly_Score'].hist(by=X['Anomaly_Class'])
```

```python

#Descriptive Statistics
X['Anomaly_Score'].describe()
```

# Local Outlier Factor

```python
#Implementing Algorithm With Different Parameters
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors=5)
clf.fit(X_normalized)
```

```python
#Predicting Anomaly Class and Anomaly Score
Anomaly_Class = clf.fit_predict(X_normalized)
Anomaly_Score = clf._decision_function(X_normalized)
```

```python
#Creating New Columns Based on Anomaly Detection
X['Anomaly_Class_neighbors_5'] = Anomaly_Class
X['Anomaly_Score_neighbors_5'] = Anomaly_Score
```

```python
#Viewing Class Counts
#-1 is anomaly, 1 is normal
X.groupby('Anomaly_Class_neighbors_5')[['Anomaly_Score_neighbors_5']].count()
```

```python
#Histogram of Anomaly Scores By Anomaly Class
X['Anomaly_Score_neighbors_5'].hist(by=X['Anomaly_Class_neighbors_5'])
```

```python
#Descriptive Statistics
X['Anomaly_Score_neighbors_5'].describe()
```

```python
#Implementing Algorithm
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(algorithm='ball_tree')
clf.fit(X_normalized)
```

```python
#Predicting Anomaly Class and Anomaly Score
Anomaly_Class = clf.fit_predict(X_normalized)
Anomaly_Score = clf._decision_function(X_normalized)
```

```python
#Creating New Columns Based on Anomaly Detection
X['Anomaly_Class_ball_tree'] = Anomaly_Class
X['Anomaly_Score_ball_tree'] = Anomaly_Score
```

```python
#Viewing Class Counts
#-1 is anomaly, 1 is normal
X.groupby('Anomaly_Class_ball_tree')[['Anomaly_Score_ball_tree']].count()
```

```python
#Histogram of Anomaly Scores By Anomaly Class
X['Anomaly_Score_ball_tree'].hist(by=X['Anomaly_Class_ball_tree'])
```
